Task_1:  Generate the random signal of size 10,000 in the range of 0 to 99.

```python
import numpy as np
import matplotlib.pyplot as plt

samples=np.random.randint(100,size=10000)
print("Number of samples:",len(samples))
print(samples)
```

Task_2: Count the occurrence of each sample from the signal. Plot the occurrence of each sample.

```python
def occurance(samples):
  ocr=np.zeros(100)
  for i in range(100):

    ocr[i]=sum(samples==i)
  return ocr
ocr=occurance(samples)

plt.plot(ocr)
plt.title(" Occurences of each sample")
plt.xlabel("Number of unique samples")
plt.ylabel("count of occurence of unique samples") 
```

Task_3: Plot the probability of each sample

```python
def probability(samples):

 prb=np.zeros(100)
 for i in range(100):
   prb[i]=sum(samples==i)/len(samples)
 return prb
prb=probability(samples)
#Plotting the probability of each sample.
plt.plot(prb)
plt.title(" PDF of Occurences")
plt.xlabel("Number of unique samples")
plt.ylabel("Probability of unique samples") 
```
